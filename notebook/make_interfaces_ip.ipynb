```python
links = open("../docs/links_infos_eve1.md").readlines()
```

```python
ips = {}

for line in links:
    if "xrv" in line and "Gi" in line:
        line = line.strip().split("|")
        ips.setdefault(line[0].strip(),[])
        ips[line[0].strip()].append({"interface_name": line[1].strip(), "ipaddress": line[2].strip().replace("/30", "")})
        
        ips.setdefault(line[-1].strip(),[])
        ips[line[-1].strip()].append({"interface_name": line[-2].strip(), "ipaddress": line[-3].strip().replace("/30", "")})

for line in links:
    if "Loopback0" in line:
        line = line.strip().split("|")

        ips[line[0].strip()].append({"interface_name": line[1].strip(), "ipaddress": line[2].strip()})
```

```python
ips
```

```python
from jinja2 import Template
template = '''
interfaces:{% for interface in interfaces %}
  - name: {{ interface.interface_name }}
    ipaddress: {{ interface.ipaddress }}{% endfor %}

'''
yaml_file = Template(template)

for hostname, interfaces in ips.items():
    if "xrv" in hostname or "vIOS" in hostname:
        interfaces_context = yaml_file.render(interfaces=interfaces)
        filename = "../data/interfaces/%s_interfaces.yaml" %(hostname)
        f = open(filename,"w")
        
        f.write(interfaces_context)
        print(interfaces_context)
        f.close()
```
