现在我们 获得了一个变量  devices_console。这个是一个列表。里面包含了所有的设备的console 和登录信息。 

继续解析这个内容

```python
devices = []
for device in devices_console:
    device.strip()     # 删除一行后最后的一个换行符号
    hostname, port, username, password, _type = device.split("|")      # type 是系统的关键字，这里在前面加一个 _ 来区分系统的关键字
    hostname = hostname.strip()
    port = port.strip()
    port = int(port)
    username = username.strip()
    password = password.strip()
    
    devices.append({"hostname": hostname,
                  "port": port,
                  "username": username,
                  "password": password})

pprint(devices)
```

## 1.2 用telnetlib 来登录设备
telnetlib 是系统内置的一个库

```python
from telnetlib import Telnet
import re
import time
```

```python

def sendcmd(tn, cmd, sleep_time=0.5):
    if isinstance(cmd, str):
        cmd = str.encode(cmd)
    tn.write(cmd + b"\n")
    time.sleep(sleep_time)
    return tn

def login(hostname, port, username="admin", password="admin", timeout=2, debug=False):

    # create a Telnet object instance
    tn = Telnet(host=hostname, port=port)
    
    if debug:
        tn.set_debuglevel(7)
    
    '''send return key 
       console 在idle 时，需要先发送一个return key 才能到登录界面。
       如:
       ios con0/0/CPU0 is now available
 
 
       Press RETURN to get started.
    '''
    tn = sendcmd(tn, "\r")

    
    for _ in range(0, 2):
        '''
        创建一个登录过程中，希望读到的字符串。在python3 的telnetlib中，需要bytes类型。
        如果看到 Username： 那么就发送 用户名。
        如果看到 Password： 就发送 密码。
        如果看到 RP/0/0/CPU0 就说明已经登录了。 在这个LAB中，所有的设备只有单引擎。因此在IOS XR 上都是 RP/0/0/CPU0
        如果看到是 Router 就说明已经登录完成了。在这个LAB中，vIOS设备在没有配置的时，hostname 默认是 Router
        
        
        '''
        expect_list = []
        expect_list.append(re.compile(b"username:", re.I))       # re.I 不区分大小写
        expect_list.append(re.compile(b"password:",re.I))        # re.I 不区分大小写
        expect_list.append(re.compile(b"RP/0/0/CPU0", re.I))     # re.I 不区分大小写
        expect_list.append(re.compile(b"Router"))                
        
        result = tn.expect(expect_list, timeout=timeout)
        if result[0] == 0:
            sendcmd(tn, username)
        elif result[0] == 1:
            sendcmd(tn, password)
        elif result[0] == 2 or result[0] == 3:
            print("Login right")
            break
        else:
            print("Login Fail")
    return tn
        


#tn = login("172.20.0.1", port=33283)

```

## 1.3 登录完成后配置设备的主机名 

```python
def config_hostname(tn, hostname):
    # 进入配置模式
    sendcmd(tn, "config terminal")
    # 修改主机名
    sendcmd(tn, "hostname %s" %hostname)
    # 在 IOS XR 平台需要commit 操作
    sendcmd(tn, "commit")

    # 退出配置模式
    sendcmd(tn, "end")
    # 关闭 telnet 的会话
    tn.close()
    
tn = login("172.20.0.1", port=33282,debug=True)    
config_hostname(tn, "xrv2")
```

## 1.4 修改mgt口的IP地址信息

```python
def config_mgt_ip(tn, ipaddress):
    # 进入配置模式
    sendcmd(tn, "config terminal")
    sendcmd(tn, "interface mgmtEth 0/0/CPU0/0")
    sendcmd(tn, "ip address %s 255.255.0.0" %ipaddress)
    sendcmd(tn, "no shutdown")    
    sendcmd(tn, "commit")
    
    # 退出配置模式
    sendcmd(tn, "end")
    # 关闭 telnet 的会话
    tn.close()
    
tn = login("172.20.0.1", port=33283)
config_mgt_ip(tn, "172.20.3.104")
```

## 2.0 通过SSH登录设备
通过SSH 登录设备，并通过ssh 获取设备的一些信息。

这次使用的lib 为 [netmiko]。(https://github.com/ktbyers/netmiko)

这是一个开源的python 第三方库。目前支持大量的网络设备。其就是操作网络设备而开发的第三方库。

### 2.1 登录SSH后，获取设备的接口信息

```python
from netmiko import ConnectHandler
```

```python
iosxr3 = {
    'device_type': 'cisco_xr',
    'ip':   '172.20.3.102',
    'username': 'admin',
    'password': 'admin',
    'port' : 22          # optional, defaults to 22    
}

net_connect = ConnectHandler(**iosxr3)

output = net_connect.send_command('show ip int brief')
print(output)
```

## 2.2 修改接口的IP地址信息

```python
interface_cfg = ["interface GigabitEthernet0/0/0/1",
                "no shutdown",
                "commit",
                "end"]

output = net_connect.send_config_set(interface_cfg)
print(output)
```
